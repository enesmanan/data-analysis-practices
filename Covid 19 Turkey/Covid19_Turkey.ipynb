```python
ay = []
for current_day in df['Date']:
    if 0 <= current_day <= 20:
        ay.append('Mart')
    elif 20 < current_day <= 50:
        ay.append('Nisan')
    elif 50 < current_day <= 80:
        ay.append('Mayıs')
    elif 80<current_day<=112:
        ay.append('Haziran')
ay = pd.DataFrame(ay)
df['Month'] = ay
print("Mart vakası: {}".format(df.iloc[20,2])) 
print("Nisan vakası: {}".format(df.iloc[50,2]-df.iloc[20,2]))
print("Mayıs vakası: {}".format(df.iloc[80,2]-df.iloc[50,2]))
print("Haziran vakası: {}".format(df.iloc[110,2]-df.iloc[80,2])) 
     
```
